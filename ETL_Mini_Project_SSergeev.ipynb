### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
columns = crowdfunding_info_df.columns
print(columns)
```

```python
# Extract the category and subcategory from the 'category & sub-category' column
crowdfunding_info_df[['category', 'subcategory']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)

# Get unique categories
categories = crowdfunding_info_df['category'].unique()

# Create category DataFrame
category_df = pd.DataFrame({
    'category_id': [f'cat{i}' for i in range(1, len(categories) + 1)],
    'category': categories
})

# Display the category DataFrame
print(category_df)
```

```python
# Get unique subcategories
subcategories = crowdfunding_info_df['subcategory'].unique()

# Create subcategory DataFrame
subcategory_df = pd.DataFrame({
    'subcategory_id': [f'subcat{i}' for i in range(1, len(subcategories) + 1)],
    'subcategory': subcategories
})

# Display the subcategory DataFrame
print(subcategory_df)
```

```python
# Get the unique categories and subcategories in separate lists.
print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Create the campaign_df DataFrame
campaign_df = pd.DataFrame({
    # Copy basic information from crowdfunding_info_df
    'cf_id': crowdfunding_info_df['cf_id'],
    'contact_id': crowdfunding_info_df['contact_id'],
    'company_name': crowdfunding_info_df['company_name'],
    
    # Rename 'blurb' to 'description'
    'description': crowdfunding_info_df['blurb'],
    
    # Convert 'goal' and 'pledged' to float type
    'goal': crowdfunding_info_df['goal'].astype(float),
    'pledged': crowdfunding_info_df['pledged'].astype(float),
    
    # Copy other basic information
    'outcome': crowdfunding_info_df['outcome'],
    'backers_count': crowdfunding_info_df['backers_count'],
    'country': crowdfunding_info_df['country'],
    'currency': crowdfunding_info_df['currency'],
    
    # Convert 'launched_at' and 'deadline' to datetime, renaming them
    'launch_date': pd.to_datetime(crowdfunding_info_df['launched_at'], unit='s', utc=True),
    'end_date': pd.to_datetime(crowdfunding_info_df['deadline'], unit='s', utc=True),
    
    # Map category and subcategory to their respective IDs
    'category_id': category_df.set_index('category')['category_id'].loc[crowdfunding_info_df['category']].values,
    'subcategory_id': subcategory_df.set_index('subcategory')['subcategory_id'].loc[crowdfunding_info_df['subcategory']].values
})

# Display the first few rows of the DataFrame
campaign_df.head()

```

```python
# Show the data types
campaign_df.info()
```

```python
# Merge the campaign_df with the category_df on the "category_id" column and 
# the subcategory_df on the "subcategory_id" column.
campaign_merged_df = pd.merge(campaign_df, category_df, on='category_id', how='left')
campaign_merged_df = pd.merge(campaign_merged_df, subcategory_df, on='subcategory_id', how='left')


campaign_merged_df.head()
```

```python
# Drop unwanted columns
campaign_cleaned = campaign_merged_df.drop(columns=['category', 'subcategory'])
# Show the data types
campaign_cleaned.head()
```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

```python
# Extract the name of the contact and add it to a new column.
contacts_info_df_copy['name'] = contacts_info_df_copy['contact_info'].apply(lambda x: json.loads(x)['name'])

# Print the contacts_info_df_copy DataFrame
contacts_info_df_copy.head()



```

```python
# Extract the email from the contacts and add the values to a new column.
contacts_info_df_copy['email'] = contacts_info_df_copy['contact_info'].apply(lambda x: json.loads(x)['email'])

# Print the contacts_info_df_copy DataFrame
contacts_info_df_copy.head()

```

```python
# Create a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
contacts_df_clean = contacts_info_df_copy[['contact_id', 'name', 'email']]

# Print the contacts_df_clean DataFrame
contacts_df_clean.head()

```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
contacts_df_clean['first_name'] = contacts_df_clean['name'].apply(lambda x: x.split()[0])
contacts_df_clean['last_name'] = contacts_df_clean['name'].apply(lambda x: x.split()[1])

# Print the contacts_df_clean DataFrame
contacts_df_clean.head()

# Drop the contact_name column
contacts_df_clean = contacts_df_clean.drop(columns=['name'])

# Print the contacts_df_clean DataFrame
contacts_df_clean.head()
```

```python
# Reorder the columns
contacts_df_clean = contacts_df_clean[['contact_id', 'first_name', 'last_name', 'email']]

# Change the datatype of the "contact_id" column to an int64 data type.
contacts_df_clean['contact_id'] = contacts_df_clean['contact_id'].astype('int64')

# Print the contacts_df_clean DataFrame
contacts_df_clean.head()

```
